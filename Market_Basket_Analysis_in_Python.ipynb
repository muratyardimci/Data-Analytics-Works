```python
import pandas as pd
```

```python
data = pd.read_excel("/content/drive/MyDrive/online_retail_II.xlsx")
```

```python
data.info()
```

```python
data = data.dropna()
data.info()
```

```python
data_plus = data[data["Quantity"] > 0]
data_plus.info() # 407695 entries: basket that our customer actually bought, not cancelled.
```

```python
basket_plus = (data_plus[data_plus['Country'] == "United Kingdom"].groupby(['Invoice', 'Description'])['Quantity']
  .sum().unstack().reset_index().fillna(0).set_index('Invoice'))
basket_plus # only the transaction that come from UK. This basket data will contain the Quantity of each items bought per transaction (Invoice)
```

```python
def encode_units(x):
  if x <= 0:
    return 0
  if x >= 1:
    return 1
# 1 is bought, 0 is not bought items
basket_encode_plus = basket_plus.applymap(encode_units)
basket_encode_plus
```

```python
#Filter The Transaction : Bought More Than 1 Items Only
basket_filter_plus = basket_encode_plus[(basket_encode_plus > 0).sum(axis=1) >= 2]
basket_filter_plus
```

```python
#Apply the Apriori Algorithm : simply used to find the frequently bought items in the dataset
!pip install mlxtend
from mlxtend.frequent_patterns import apriori
```

```python
frequent_itemsets_plus = apriori(basket_filter_plus, min_support=0.03, use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)
frequent_itemsets_plus['length'] = frequent_itemsets_plus['itemsets'].apply(lambda x: len(x))
frequent_itemsets_plus
```

```python
# Finding The Association Between Frequently Bought Items
from mlxtend.frequent_patterns import association_rules

# Get the total number of transactions from the original filtered basket data
num_transactions = len(basket_filter_plus)

# Apply association rules with the num_itemsets argument
rules = association_rules(frequent_itemsets_plus,
                         metric="lift",
                         min_threshold=1,
                         num_itemsets=num_transactions)  # Pass num_transactions here

# Sort and reset index
rules = rules.sort_values('lift', ascending=False).reset_index(drop=True)
rules

# The higher the lift value, the higher the association between the items will.
# If the lift value is more than 1, it is enough for us to say that those two items are associated each other
```
