```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/Electric_Vehicle_Population_Data.csv")
```

```python
df.info()
```

```python
df.tail(5)
```

```python
make = df.Make[:350]
```

```python
value_make = pd.DataFrame(make.value_counts())
```

```python
value_make
```

```python
plt.plot(value_make)
```

```python
plt.bar(value_make.index, value_make['count']);
```

```python
plt.bar(value_make.index, value_make['count'], color='red');
plt.xticks(rotation=45, ha= 'right');
```

```python
plt.barh(value_make.index, value_make['count'], color='red');
```

```python
model_year = df['Model Year'][:350]
plt.plot(model_year);
```

```python
model_year = df['Model Year'][:350]
plt.hist(model_year);
```

```python
plt.hist(df['Model Year'])
```

```python
Company = pd.DataFrame(df.groupby("Make")["Electric Range"].mean().sort_values(ascending=False)).head(15)
Company.head()
```

```python
plt.figure(figsize=[14,6])
plt.bar(Company.index, Company['Electric Range'])
plt.xticks(rotation=45, ha='right')
plt.title("Average Electric Range");
```

```python
plt.figure(figsize =[7,7])
plt.boxplot(df['Model Year']);
```
