```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import seaborn as sns
```

```python
housing = datasets.fetch_california_housing()
dataset = pd.DataFrame(housing.data, columns=housing.feature_names)
dataset['median_house_value'] = housing.target
```

```python
dataset.info()
```

```python
dataset.head(5)
```

```python
sns.histplot(dataset['HouseAge'], bins=8, kde=True)
plt.xlabel('House Age')
plt.ylabel('Frequency')
plt.title('House Age Distribution')
plt.show()
```

```python
sns.jointplot(x='AveRooms', y='median_house_value', data=dataset, kind='reg')
plt.xlabel('House Age')
plt.ylabel('Target')
plt.title('House Age vs. Median House Value')
plt.show()
```

```python
sns.heatmap(dataset.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(dataset)
plt.show()
```

```python
sns.histplot(dataset['MedInc'],bins=30, kde = True)
plt.xlabel('Median Income')
plt.ylabel('Frequency')
plt.title('Median Income Distribution')
plt.show()
```

```python
sns.pairplot(dataset, x_vars=["AveBedrms", "AveRooms"], y_vars=["median_house_value"],
             height=5, aspect=.8, kind="reg");
```
